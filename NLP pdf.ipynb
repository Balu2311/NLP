```python
import nltk
import PyPDF2
import os
import re
```

```python
Data = open("D:\\WorkSpace\\M L P\\Supervised Learning\\NLP\\Hindu_Marriage_Act.pdf", "rb")
```

```python
Freader = PyPDF2.PdfFileReader(Data)
Freader.numPages
```

```python
File = Freader.getPage(0)
PDF = File.extractText()
```

```python
PDF
```

```python
pattern = re.compile(r"$")
print(pattern)
```

```python
d1 = pattern.finditer(PDF)
for i in (list(d1)):
    print(i)
```

```python
m = re.match(r".?", PDF)
print(m)
if m:
    print("First :", m[0])
else:
    print("no match")
```

```python
s = re.search(r" [A].* \d\d off? \d{4}", PDF)
print(s)
if s:
    print("search:", s[0])
else:
    print("No match")
```

```python
f = re.findall(r"\d{4}.*", PDF)
f
```

```python
PDF
```

```python
Special = re.sub(r"\W", " ", PDF)
Special
```

```python
space = re.sub(" +", " ", Special)
```

```python
space
```

```python
from nltk.tokenize import word_tokenize
```

```python
WORD = word_tokenize(space)
print(WORD)
```

```python
from nltk.tokenize import sent_tokenize
```

```python
sentences = sent_tokenize(space)
sentences
```

```python
print(WORD)
```

```python
NumR = re.sub(r"[^a-zA-Z]", " ", PDF)
NumR
```

```python
Sp = re.sub(r" +", " ", NumR)
Sp
```

```python
Wo = word_tokenize(Sp)
Wo
```

```python
Lo = (x.lower() for x in Wo)
a =(list(Lo))
print(a)
```

```python
len(a)
```

```python
b = sorted(set(a))
b
```

```python
len(b)
```

```python
pos = nltk.pos_tag(b)
pos
```

```python
from nltk.corpus import stopwords
```

```python
STOP = stopwords.words("english")
STOP
```

```python
len(STOP)
```

```python
clean = [x for x in b if x not in STOP]
clean
```

```python
len(clean)
```

```python
from nltk.util import bigrams,trigrams,ngrams
```

```python
Bigrams = list(bigrams(clean))
Bigrams
```

```python
Trigrams = list(trigrams(clean))
Trigrams
```

```python
Ngrams = list(ngrams(clean, 5))
Ngrams
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps= PorterStemmer()
Stem_Words=[ps.stem(x) for x in clean]
print(clean)
print(Stem_Words)
```

```python
from nltk.stem.lancaster import LancasterStemmer
```

```python
ls = LancasterStemmer()
Lan_Words=[ls.stem(x) for x in clean]
print(Lan_Words)
print(clean)
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
wl = WordNetLemmatizer()
Lemma_Word=[wl.lemmatize(x) for x in clean]
print(Lemma_Word)
print(clean)
```

```python
print(wl.lemmatize('better', pos="a"))
print(wl.lemmatize('asked', pos="v"))
print(wl.lemmatize('slowly', pos="r"))
```
